```python
driver=webdriver.Chrome(r'Chromedriver.exe')
```

```python
driver.maximize_window()
time.sleep(1)
```

```python
driver.get('https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1')
```

```python
Movie_urls=[]

start=0
end=50

for page in range(start,end):
    
    try:
        url=driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"][1]/a[1]')
        if len(url)>0:
            for i in url:
                Movie_urls.append(i.get_attribute('href'))
        else:
            Movie_urls.append('-')
        next_button=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]') 
        next_button.click()
        time.sleep(2)
    except NoSuchElementException:
        Movie_urls.append('no data')
        
print(len(Movie_urls))
```

```python
Movie=[]
for url in Movie_urls[0:2]:
    driver.get(url)
    time.sleep(1)
    
    try:
        movie=driver.find_elements(By.XPATH,'//span[@class="lister-item-index unbold text-primary"]')
        for i in movie:
            Movie.append(i.text)
            
    except NoSuchElementException:
        Movie.append('-')
```

```python
Names=[]
Duration=[]
MetaScore=[]
Director1=[]
Director2=[]
Ratings=[]
Star1=[]
Star2=[]
Star3=[]
Star4=[]
Genre1=[]
Genre2=[]
Genre3=[]
Year=[]
Papularity=[]
Certificates=[]


for url in Movie_urls[0:1770]:
    driver.get(url)
    time.sleep(2)
    
    try:
        names=driver.find_elements(By.XPATH,'//span[@class="sc-afe43def-1 fDTGTb"]')
        if len(names)>0:
            for i in names:
                Names.append(i.text)
        else:
            Names.append('-')
    except NoSuchElementException:
        Names.append('no data found')
        
        
    try:
        ratings=driver.find_elements(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/div[1]/div/div[1]/a/span/div/div[2]/div[1]/span[1]')
        if len(ratings)>0:
            for i in ratings:
                Ratings.append(i.text)
        else:
            Ratings.append('-')
        time.sleep(2)
    except NoSuchElementException:
        Ratings.append('no data found')
    except StaleElementReferenceException:
        Ratings.append('no data')
        
    try:
        director1=driver.find_elements(By.XPATH,'/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/section[4]/ul/li[1]/div/ul/li[1]/a')
        if len(director1)>0:
            for i in director1:
                Director1.append(i.text)
        else:
            Director1.append('-')
        time.sleep(2)
    except NoSuchElementException:
        Director1.append('No Data')
    except StaleElementReferenceException:
        Director1.append('Not present')
        
    try:
        director2=driver.find_elements(By.XPATH,'/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/section[4]/ul/li[1]/div/ul/li[2]/a')
        if len(director2)>0:
            for i in director2:
                Director2.append(i.text)
        else:
            Director2.append('-')
        time.sleep(2)
    except NoSuchElementException:
        Director2.append('No Data')
    except StaleElementReferenceException:
        Director2.append('Not present')
    
    try:
        metascore=driver.find_elements(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/ul/li[3]/a/span/span[1]/span')
        if len(metascore)>0:
            for i in metascore:
                MetaScore.append(i.text)
        else:
            MetaScore.append('-')
        time.sleep(2)
        
    except NoSuchElementException:
        metascore=driver.find_elements(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/ul/li[3]/a/span/span[1]/span')
        MetaScore.append('')
    except StaleElementReferenceException:
        MetaScore.append('not found')
        
    try:
        star1=driver.find_elements(By.XPATH,'/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/section[4]/div[2]/div[2]/div[1]/div[2]/a')
        if len(star1)>0:
            for i in star1:
                Star1.append(i.text)
        else:
            Star1.append('-')
        time.sleep(2)
    except NoSuchElementException:
        Star1.append('-')
    
    try:
        star2=driver.find_elements(By.XPATH,'/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/section[4]/div[2]/div[2]/div[2]/div[2]/a')
        if len(star2)>0:
            for i in star2:
                Star2.append(i.text)
        else:
            Star2.append('-')
        time.sleep(2)
    except NoSuchElementException:
        Star2.append('-')
        
    try:
        star3=driver.find_elements(By.XPATH,'/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/section[4]/div[2]/div[2]/div[3]/div[2]/a')
        if len(star3)>0:
            for i in star3:
                Star3.append(i.text)
        else:
            Star3.append('-')
        time.sleep(2)
    except NoSuchElementException:
        Star3.append('-')
        
    try:
        star4=driver.find_elements(By.XPATH,'/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/section[4]/div[2]/div[2]/div[4]/div[2]/a')
        if len(star4)>0:
            for i in star4:
                Star4.append(i.text)
        else:
            Star4.append('-')
        time.sleep(2)
    except NoSuchElementException:
        Star4.append('-')
        
    try:
        genre1=driver.find_elements(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]/a[1]')
        if len(genre1)>0:
            for i in genre1:
                Genre1.append(i.text)
        else:
            Genre1.append('-')
        time.sleep(2)
    except NoSuchElementException:
        Genre1.append('-')
        
    try:
        genre2=driver.find_elements(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]/a[2]')
        if len(Genre2)>0:
            for i in genre2:
                Genre2.append(i.text)
        else:
            Genre2.append('-')
        time.sleep(2)
    except NoSuchElementException:
        Genre2.append('-')
        
    try:
        genre3=driver.find_elements(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]/a[3]')
        if len(genre3)>0:
            for i in genre3:
                Genre3.append(i.text)
        else:
            Genre3.append('-')
        time.sleep(2)
    except NoSuchElementException:
        Genre3.append('-')
        
    try:
        year=driver.find_elements(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[1]/a')
        if len(year)>0:
            for i in year:
                Year.append(i.text)
        else:
            Year.append('-')
        time.sleep(2)
    except NoSuchElementException:
        Year.append('-')
    except StaleElementReferenceException:
        Year.append('-')   
        
    try:
        papularity=driver.find_elements(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/div[1]/div/div[3]/a/span/div/div[2]/div[1]')
        if len(papularity)>0:
            for i in papularity:
                Papularity.append(i.text)
        else:
            Papularity.append('-')
        time.sleep(2)
    except NoSuchElementException:
        Papularity.append('-')
    except StaleElementReferenceException:
        Papularity.append('-')
        
    try:
        certificates=driver.find_elements(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[2]/a')
        if len(certificates)>0:
            for i in certificates:
                Certificates.append(i.text)
        else:
            Certificates.append('-')
        time.sleep(2)
    except NoSuchElementException:
        Certificates.append('-')
    except StaleElementReferenceException:
        Certificates.append('-')
        
    try:
        duration=driver.find_elements(By.XPATH,'//ul[@class="ipc-inline-list ipc-inline-list--show-dividers sc-afe43def-4 kdXikI baseAlt"][1]/li[3]')
        if len(duration)>0:
            for i in duration:
                Duration.append(i.text)
        else:
            Duration.append('-')
            
        time.sleep(3)
    except NoSuchElementException:
        Duration.append('no data found')
        
print('Duration',len(Duration))

    
print('Year',len(Year))  

    
print('Papularity_score',len(Papularity))

        
print('Genre 1',len(Genre1))

print('Genre 2',len(Genre2))

print('Genre 3',len(Genre3))


        
print('Star 1',len(Star1))

print('Star 2',len(Star2))

print('Star 3',len(Star3))

print('Star 4',len(Star4))

        
print('MetaScore',len(MetaScore))


print('Director1',len(Director1))

        
print('Director2',len(Director2))

        
print('Ratings',len(Ratings))
  
        
print('Names',len(Names))
```
