# Importing Csv file data into sql

```python
import csv 
import sqlite3 

db = sqlite3.connect("D://My sqlite/Student_database.db") 
cursor = db.cursor() 
cursor.execute("create Table IMDB_Movie_1(MovieName TEXT PRIMARY KEY,Director1_Name TEXT,Director2_Name TEXT,Duration TEXT,Genre_1 TEXT,Genre_3 TEXT,Ratings FLOAT)")

with open('IMDB_MOVIES_LIST_1', encoding='utf-8') as file:
    total_records = 0
    csv_data = csv.reader(file)
    next(csv_data) 
    for row in csv_data:
        try:
            cursor.execute("insert into IMDB_Movie_1 values(?,?,?,?,?,?,?)", row)
            db.commit()
            total_records += 1
        except sqlite3.IntegrityError:
            print(f"Record {row} already exists")
            continue

print(total_records,'records inserted')
```

```python
#Query No =1
result=cursor.execute("SELECT * FROM IMDB_Movie_1 WHERE Director1_Name IN ('Christopher', 'Matt Reeves')")
for i in result:
    print(i)
```

```python
#Query No = 2
result=cursor.execute("SELECT * FROM IMDB_Movie_1 WHERE CAST(SUBSTR(duration, 1, INSTR(Duration, 'h') - 1) AS INTEGER) * 60 + CAST(SUBSTR(Duration, INSTR(duration, 'h') + 1, INSTR(duration, 'm') - INSTR(duration, 'h') - 1) AS INTEGER) BETWEEN 140 AND 190")
for i in result:
    print(i)
```

```python
#Query No =3
result=cursor.execute('select * from IMDB_Movie_1 where Ratings>7 ')
for i in result:
    print(i)
```

```python
#Query No = 4

result=cursor.execute('select * from IMDB_Movie_1 where MovieName like "p%" AND Ratings>7 ')
for i in result:
    print(i)
```

# CSV File NO - 2

```python
import sqlite3 

db = sqlite3.connect("D://My sqlite/Student_database.db") 
cursor = db.cursor() 
cursor.execute('create Table IMDB_Movie_2(Movie_Name text primary key,Star_1 text,Star_2 text,Star_3 text,Star_4 text,Genre_1 text,Genre_3 text,Papularity text,Certification text)')

with open('IMDB_MOVIES_LIST_2', encoding='utf-8') as file:
    total_records = 0
    csv_data = csv.reader(file)
    next(csv_data) # skip header row
    for row in csv_data:
        try:
            cursor.execute("insert into IMDB_Movie_2 values(?,?,?,?,?,?,?,?,?)", row)
            db.commit()
            total_records += 1
        except sqlite3.IntegrityError:
            print(f"Record {row} already exists")

print(total_records,'records inserted')
```

```python
result=cursor.execute("SELECT Movie_Name FROM IMDB_Movie_2 WHERE Star_1 = 'Arnold Schwarzenegger'")
for i in result:
    print(i)
```
